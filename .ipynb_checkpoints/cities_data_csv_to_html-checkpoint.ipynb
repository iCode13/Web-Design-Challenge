```python
# Dependencies
import pandas as pd
from datetime import datetime

# Read CSV file into a dataframe
cities_df = pd.read_csv("Resources/cities.csv")
cities_df.head()
```

```python
# Reorder columns
cities_df = cities_df[["City", "Country", "Date", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed"]]

# Format date
cities_df["Date"] = pd.to_datetime(cities_df['Date'], unit='s')

# Format column headers
cities_df["Lat"] = cities_df["Lat"].replace("Lat", "Latitude")
cities_df["Lng"] = cities_df["Lng"].replace("Lng", "Longitude")
cities_df.rename(columns = {"City":"City",  "Country":"Country", "Date":"Date", "Lat":"Latitude", "Lng":"Longitude", "Max Temp":"Maximum Temperature (F)",
                          "Cloudiness":"Cloudiness", "Wind Speed":"Wind Speed"}, inplace = True) 

cities_df
```

```python
# Export dataframe as HTML
cities_df.to_html("Web-Visualizations/cities_data.html")
```
